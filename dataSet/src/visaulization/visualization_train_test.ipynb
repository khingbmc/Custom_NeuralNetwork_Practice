```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
import pygal
import seaborn as sns
```

```python
f = open('../test_iris/test/testing0.txt', 'r')
x = f.readlines()
f.close()
f = open('../test_iris/train/training0.txt', 'r')
y = f.readlines()
f.close()
```

```python
test_data = np.array(ast.literal_eval(x[0]))
train_data = np.array(ast.literal_eval(y[0]))
```

```python
df_train = pd.DataFrame(train_data, columns=['x', 'y', 'z', 'w', 'class'])
df_test = pd.DataFrame(test_data, columns=['x', 'y', 'z', 'w', 'class'])
```

```python
fig = plt.figure("plot train test x y")
xy = fig.add_axes([0,0,1,1])
xy.scatter(df_train[df_train['class']==0]['x'], df_train[df_train['class']==0]['y'], color='r', s=20, marker='^')
xy.scatter(df_test[df_test['class']==0]['x'], df_test[df_test['class']==0]['y'], color='r', s=60, marker='o')
xy.scatter(df_train[df_train['class']==1]['x'], df_train[df_train['class']==1]['y'], color='g', s=20, marker='^')
xy.scatter(df_test[df_test['class']==1]['x'], df_test[df_test['class']==1]['y'], color='g', s=60, marker='o')
xy.scatter(df_train[df_train['class']==2]['x'], df_train[df_train['class']==2]['y'], color='b', s=20, marker='^')
xy.scatter(df_test[df_test['class']==2]['x'], df_test[df_test['class']==2]['y'], color='b', s=60, marker='o')
xy.set_xlabel('x')
xy.set_ylabel('y')
xy.set_title('plot x y')
plt.savefig('xy0.png')
```
